```python
#Task 7

class Cat: 
    Number_of_cats = 0   
    def __init__(self, color = "Default Color", doing = "Nothing"):
        self.color = color 
        self.doing = doing
        Cat.Number_of_cats = Cat.Number_of_cats + 1

    @classmethod
    def no_parameter (cls, color="White", doing = "sitting"):
        return cls(color, doing) 

    @classmethod
    def first_parameter (cls, color="", doing = "sitting"):
        return cls(color, doing) 

    @classmethod
    def second_parameter (cls, color="Grey", doing = "playing"):
        color = "Grey"
        doing = "Playing"
        return cls(color, doing) 

    def changeColor (self, color):
        self.color = color 

    def printCat (self):
        print (self.color, " cat is ", self.doing)


print("Total number of cats:", Cat.Number_of_cats) 
c1 = Cat.no_parameter() 
c2 = Cat.first_parameter("Black") 
c3 = Cat("Brown", "jumping") 
c4 = Cat("Red", "purring") 
c5 = Cat.second_parameter("playing")
print("=======================") 
c1.printCat() 
c2.printCat() 
c3.printCat() 
c4.printCat() 
c5.printCat() 
c1.changeColor("Blue") 
c3.changeColor("Purple") 
c1.printCat() 
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8

class Cylinder:
    radius = 5  
    height = 18

    def __init__(self, radius = 0, height = 0):
        self.radius = radius 
        self.height = height
        print("Default radius= ", Cylinder.radius, " and ", "height= ", Cylinder.height) 
        print("Updated radius= ", self.radius, " and ", "height= ", self.height) 
        Cylinder.radius = self.radius
        Cylinder.height = self.height

    @staticmethod
    def volume (radiusStatic, heightstatic):
        ans = 3.1416*(radiusStatic*radiusStatic)*heightstatic
        print ("Volume: ", ans)

    @staticmethod
    def area (radiusStatic, heightstatic):
        ans = 2*3.1416*(radiusStatic*radiusStatic) + 2*3.1416*radiusStatic*heightstatic
        print ("Area: ", ans)

    @classmethod
    def swap (cls, radius=0, height = 0):
        return cls(height, radius) 

    @classmethod
    def changeFormat (cls, theText):
        x = theText.split("-")
        return cls(float(x[0]), float(x[1])) 



c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
