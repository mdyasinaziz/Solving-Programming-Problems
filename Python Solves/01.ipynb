```python
#String Problems #Task-6

def password_check(password):
    special_sym = ['$', '@', '#', '_']
    pass_check = True

    if not any(char.isdigit() for char in password):
        print('Digit missing')
        pass_check = False

    if not any(char.isupper() for char in password):
        print('Uppercase letter missing')
        pass_check = False

    if not any(char.islower() for char in password):
        print('Lowercase letter missing')
        pass_check = False

    if not any(char in special_sym for char in password):
        print('Special symbol missing')
        pass_check = False
    if pass_check:
        return pass_check

password = input("Enter Password: ")

if (password_check(password)):
    print("OK")
```

```python
#List Problems #Task-4

def isUbjumper(li):
    length = len(li)
    empty_list = []


    for repeat in range(length):
        empty_list.append(0)

    for repeat in range(length):
        if (repeat == 0):
            continue
        diff = abs(li[repeat - 1] - li[repeat])

        try:
            empty_list[diff] = 1
        except IndexError:
            return 0

    for repeat in range(length):
        if (repeat == 0):
            continue
        if (empty_list[repeat] != 1):
            return 0

    return 1

while (True):
    try:
        our_input = [int(item) for item in input("Give List: ").split()]
        result = isUbjumper(our_input)
        if (result == 0):
            print("Not UB Jumper")
        else:
            print("UB Jumper")
    except ValueError:
        break
```
