```python
#Task-1

class Calculator:
    val1 = 0
    val2 = 0

    @staticmethod
    def __init__():
        print("Let's calculate!")

    def add(self):
        return self.val1 + self.val2

    def subtract(self):
        return self.val1 - self.val2

    def multiply(self):
        return self.val1 * self.val2

    def divide(self):
        return self.val1 / self.val2


calc = Calculator()
funcs = {'+': calc.add, '-': calc.subtract, '*': calc.multiply, '/': calc.divide}
val1 = input()
op = input()
val2 = input()
calc.val1 = int(val1)
calc.val2 = int(val2)
func = funcs[op]
print(f'Value 1: {val1}')
print(f'Operator: {op}')
print(f'Value 2: {val2}')
print(f'Result: {func()}')
```

```python
#Task-2

class Customer:
    def __init__(self, name):
        self.name = name

    @staticmethod
    def greet(name=None):
        name_ = '' if name is None else f' {name}'
        print(f'Hello{name_}!')

    def purchase(self, *items):
        item_count = len(items)
        print(f'{self.name}, you purchased {item_count} item(s):')
        for item_name in items:
            print(item_name)


customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#Task-3

class Panda:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age

    def sleep(self, time=0):
        if time == 0:
            text = f'{self.name}\'s diration is unknown thus should have only bamboo leaves'
        else:
            if 3 <= time <= 5:
                food = 'Mixed Veggies'
            elif 6 <= time <= 8:
                food = 'Eggplant & Tofu'
            elif 9 <= time <= 11:
                food = 'Broccoli Chicken'
            text = f'{self.name} sleeps {time} hours daily and should have {food}'
        return text


panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female", 3)
panda3 = Panda("Ming Ming", "Female", 8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name, panda1.gender, panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name, panda2.gender, panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name, panda3.gender, panda3.age))

print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#Task-6

class Vehicle:
    x = 0
    y = 0

    def print_position(self):
        print(f'({self.x},{self.y})')

    def moveUp(self):
        self.y += 1

    def moveLeft(self):
        self.x -= 1

    def moveDown(self):
        self.y -= 1

    def moveRight(self):
        self.x += 1


car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#Task-8

class Student:
    def __init__(self, name="default employee", id = 0, dept = "CSE", sugg = "Default Suggesst."):
        self.name = name
        self.id = id
        self.dept = dept
        self.sugg = sugg

    def dailyEffort (self, effort = 0):
        self.effort = effort
        if effort <= 4 and effort > 2:
            self.sugg = "Suggestion: Keep up the good work!"
        elif effort <= 2:
            self.sugg = "Suggestion: Should give more effort!"
        else:
            self.sugg = "Suggestion: Excellent! Now motivate others."

    def printDetails (self):
        print ("Name: ", self.name)
        print ("ID: ", self.id)
        print ("Department: ", self.dept)
        print ("Daily Effort: ", self.effort, " Hour(s)")
        print (self.sugg)

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#Task-11

class Shinobi:
    def __init__(self, name="default name", rank = "Empty", number = 0, salary = 0):
        self.name = name
        self.rank = rank
        self.number = number
        self.salary = salary

    def calSalary (self, salary = 0):
        self.number = salary
        if self.rank == "Genin":
            self.salary = salary * 50
        elif self.rank == "Chunin":
            self.salary = salary * 100
        else:
            self.salary = salary * 500

    def changeRank(self, change = "no change"):
        self.rank = change

    def printInfo (self):
        print ("Name: ", self.name)
        print ("Rank: ", self.rank)
        print ("Number of missions: ", self.number)
        print ("Salary: ", self.salary)


naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
#Task-12

class ParcelKoro:
    def __init__(self, name="No name set", product_weight = 0, location = "nai", fee = 0, location_charge = 0):
        self.name = name
        self.product_weight = product_weight
        self.location = location
        self.fee = fee
        self.location_charge = location_charge

    def calculateFee(self, location = "nai"):
        self.location = location
        if location == "nai":
            self.location_charge = 50
        else:
             self.location_charge = 100

        if self.product_weight != 0:
            self.fee = (self.product_weight * 20) + self.location_charge

    def printDetails(self):
        print ("Customer Name: ", self.name)
        print ("Product Weight: ", self.product_weight)
        print ("Total Fee: ", self.fee)

print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
#Task-13

class Batsman:
    def __init__(self, *args):
        if len(args) >= 3:
            if isinstance(args[0], str):
                self.name = args [0]
            if isinstance(args[1], int):
                self.run = args [1]
            if isinstance(args[2], int):
                self.ball = args [2]
        else:
            if isinstance(args[0], int):
                self.run = args [0]
            if isinstance(args[1], int):
                self.ball = args [1]
                self.name = "New Batsman"

    def battingStrikeRate(self):
        self.strike = (self.run / self.ball) * 100
        return self.strike

    def printCareerStatistics(self):
        print ("Name: ", self.name)
        print ("Runs Scored: ", self.run, " ,", "Balls Faced: ", self.ball)

    def setName(self, name = "New Batsman"):
        self.name = name

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#Task-15

class Account:
    def __init__(self, name="Default Account", balance = 0, draw = 0, evalutaion = 0):
        self.name = name
        self.balance = balance
        self.draw = 0
        self.evalutaion = 0

    def details(self):
        print (self.name)
        print (self.balance)
        return ""

    def withdraw (self, draw = 0):
        if self.balance == 0 or draw > self.balance:
            print ("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        if self.balance > 0:
            self.draw = draw
            self.evalutaion = (100 * self.draw) / self.balance
            if self.evalutaion < 69.3:
                self.balance = self.balance - self.draw
                print ("Withdraw successful! New balance is: ", self.balance)
            else:
                print ("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```
